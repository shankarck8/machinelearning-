```python
movies_title=pd.read_csv("u.item",sep="\|", header=None)
```

```python
movies_title=movies_title[[0,1]]
movies_title.columns=["itemid","title"]
movies_title.head()
```

```python
df=pd.merge(df,movies_title,on="itemid")
```

```python
df.head()
df.tail()
```

```python
ratings=pd.DataFrame(df.groupby('title').mean()['rating'])
```

```python
ratings.head()
```

```python
ratings['num of ratings']=pd.DataFrame(df.groupby('title').count()['rating'])
```

create the recommender system


```python
df.head()
```

```python
movmat=df.pivot_table(index="userid",columns="title",values="rating")
```

movmat.head()

```python
movmat.head()
```

```python
starwars_user_ratings=movmat['Star Wars (1977)']
```

```python
starwars_user_ratings.head(25)
```

```python
similar_to_starwars=movmat.corrwith(starwars_user_ratings)
```

```python
similar_to_starwars
```

```python
corr_starwars=pd.DataFrame(similar_to_starwars,columns=['correlation'])
```

```python
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars
```

```python
corr_starwars.head()
```

```python
corr_starwars.sort_values('correlation',ascending=False).head(10)
```

```python
ratings
```

```python
corr_starwars=corr_starwars.join(ratings) 
```

```python
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('correlation',ascending=False).head(5)
```

```python
def pr_mov(movie_name):
    movie_user_ratings=movmat[movie_name]
    similar_to_movie=movmat.corrwith(movie_user_ratings)
    corr_movie=pd.DataFrame(similar_to_movie,columns=['correlation'])
    corr_movie.dropna(inplace=True)
    corr_movie=corr_movie.join(ratings['num of ratings'])
    predictions=corr_movie[corr_movie['num of ratings']>100].sort_values('correlation',ascending=False)
    
    return predictions
```

```python
predict_my_movie=pr_mov("True Lies (1994)")
```
